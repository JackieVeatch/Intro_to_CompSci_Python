## 3. Loops

Loops repeat actions.

A `for` loop is useful when you want to go through a list or repeat a set number of times.

```python
fruits = ["apple", "banana", "mango", "grape"]

for fruit in fruits:
    print("I like", fruit)
```

```python
for number in range(5):
    print("Number:", number)
```

A `while` loop keeps going **while** a condition is true.

```python
count = 1

while count <= 5:
    print("Count is", count)
    count += 1
```

## 4. Plotting with matplotlib

Matplotlib helps us create graphs.

We'll start by importing it.

```python
import matplotlib.pyplot as plt
```

### A simple line plot

```python
days = [1, 2, 3, 4, 5]
steps = [4000, 5500, 6000, 7000, 6500]

plt.figure(figsize=(8, 4))
plt.plot(days, steps, marker="o")
plt.title("Steps Walked Each Day")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.grid(True)
plt.show()
```

### A bar chart

```python
snacks = ["Chips", "Fruit", "Cookies", "Popcorn"]
votes = [5, 8, 6, 7]

plt.figure(figsize=(8, 4))
plt.bar(snacks, votes)
plt.title("Favorite Snack Vote")
plt.xlabel("Snack")
plt.ylabel("Votes")
plt.show()
```

Imagine a roller coaster moving through different parts of a ride.

```python
track_section = [1, 2, 3, 4, 5, 6, 7]

# Quadratic function: y = -2x^2 + 16x - 10
speed = []
for x in track_section:
    y = -2*x**2 + 16*x - 10
    speed.append(y)

plt.figure(figsize=(8, 4))
plt.plot(track_section, speed, marker="o")
plt.title("Roller Coaster Speed (Quadratic)")
plt.xlabel("Track Section")
plt.ylabel("Speed")
plt.grid(True)
plt.show()
```

## 7. Simple game: Guess the secret number

This cell uses `input()`. Run it and play.

```python
secret_number = 7
guess = int(input("Guess the secret number (1 to 10): "))

if guess == secret_number:
    print("You got it!")
elif guess < secret_number:
    print("Too low!")
else:
    print("Too high!")
```

```python
import random

secret_number = random.randint(1, 10)
guessed_correctly = False

while not guessed_correctly:
    guess = int(input("Guess the secret number (1 to 10): "))

    if guess == secret_number:
        print("Correct!")
        guessed_correctly = True
    elif guess < secret_number:
        print("Too low. Try again.")
    else:
        print("Too high. Try again.")
```

## 8. Simple game: Rock, Paper, Scissors

```python
import random

choices = ["rock", "paper", "scissors"]
computer = random.choice(choices)
player = input("Choose rock, paper, or scissors: ").lower()

print("Computer chose:", computer)

if player == computer:
    print("Tie!")
elif (player == "rock" and computer == "scissors") or \
     (player == "paper" and computer == "rock") or \
     (player == "scissors" and computer == "paper"):
    print("You win!")
elif player in choices:
    print("Computer wins!")
else:
    print("That was not a valid choice.")
```

### Option 1: Create a game where a player explores a map to find hidden treasure.
*   The player moves along a path (like positions 1 → 10)
*   A treasure is hidden at a random location
*   The player moves step-by-step trying to find it
*   Along the way, they get clues!

Below is your starter code.

```python
import random

treasure = random.randint(1, 10)
moves = []

while True:
    guess = int(input("Choose a position (1–10): "))
    moves.append(guess)

    if guess == treasure:
        print("💰 You found the treasure!")
        break
    elif abs(guess - treasure) <= 2:
        print("🔥 Warm!")
    else:
        print("❄️ Cold!")

print("Your path:", moves)
```

### Option 2: Build a reation-time game
*   The program says: “Wait for it…”
*   After a random delay, it says: “GO!”
*   The player must press Enter as fast as possible
*   The program measures their reaction time

Below is your starter code:

```python
import time
import random

reaction_times = []

print("Get ready... Wait for 'GO!' and press Enter as fast as you can!")

for round_num in range(3):
    input("\nPress Enter to start...")

    delay = random.uniform(2, 5)
    print("Wait for it...")
    time.sleep(delay)

    start_time = time.time()
    input("GO! Press Enter NOW!")
    end_time = time.time()

    reaction_time = end_time - start_time
    reaction_times.append(reaction_time)

    print(f"Your reaction time: {reaction_time:.3f} seconds")

# Results
avg_time = sum(reaction_times) / len(reaction_times)
print("\nYour times:", reaction_times)
print(f"Average reaction time: {avg_time:.3f} seconds")
```

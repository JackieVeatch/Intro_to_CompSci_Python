## 🍬 Part 1 — 1D Arrays (Candy Counts)
Imagine you surveyed 12 houses. Each value is the number of candies given at each house.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
candy_counts = np.array([2, 3, 5, 1, 4, 6, 0, 8, 10, 3, 2, 7])
candy_counts
```

**Basic stats**

```python
print("Total candy:", candy_counts.sum())
print("Average per house:", candy_counts.mean())
print("Median:", np.median(candy_counts))
print("Std dev:", candy_counts.std())
print("Max, Min:", candy_counts.max(), candy_counts.min())
```

**Indexing & slicing**

```python
# first 5 houses
first_five = candy_counts[:5]
last_three = candy_counts[-3:]
every_other = candy_counts[::2]
first_five, last_three, every_other
```

**Boolean masking** — Which houses gave at least 5 candies?

```python
mask_5plus = candy_counts >= 5
houses_5plus = np.where(mask_5plus)[0]
candy_counts[mask_5plus], houses_5plus
```

**Plot: Candy per house**

```python
plt.figure(figsize=(8,4))
plt.plot(np.arange(1, len(candy_counts)+1), candy_counts, marker='o')
plt.title("Candy per House — Trick-or-Treat Survey")
plt.xlabel("House #")
plt.ylabel("Candies")
plt.grid(True)
plt.show()
```

## 🏚️ Part 2 — 2D Arrays (Haunted House Heatmap)
Consider a 6×8 grid of rooms in a haunted house. Each value is the 'spookiness level' (0–10).

```python
np.random.seed(31)
spookiness = np.random.randint(0, 11, size=(6, 8))
spookiness
```

**2D indexing**

```python
# room at row 2, col 5 (0-indexed) 
spookiness[2, 5]
```

**Row/column slices**

```python
row_0 = spookiness[0, :]
col_3 = spookiness[:, 3]
row_0, col_3
```

**Summary by axes**

```python
print("Mean spookiness overall:", spookiness.mean())
print("Mean by row:", spookiness.mean(axis=1))
print("Mean by col:", spookiness.mean(axis=0))
```

**Masking** — Mark *very spooky* rooms (≥ 8)

```python
very_spooky = spookiness >= 8
very_spooky.astype(int)  # 1 = spooky, 0 = not
```

**Visualize the haunted house (heatmap)**

```python
plt.figure(figsize=(6,4))
plt.imshow(spookiness, aspect='auto')
plt.colorbar(label="Spookiness level (0–10)")
plt.title("Haunted House Heatmap")
plt.xlabel("Column")
plt.ylabel("Row")
plt.show()
```

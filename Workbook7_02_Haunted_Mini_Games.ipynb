# 🎃 Haunted Mini‑Games in Python
*Date: October 31, 2025*

We'll build a few tiny console games:
1. **Pumpkin Guess** — number guessing
2. **Haunted Door** — random choice
3. **Mini Text Adventure** — branching with conditionals

**Skills**: input(), print(), variables, random numbers, while loops, if/elif/else.


```python
import random
```

## Game 1 — *Pumpkin Guess* 🎃
The computer picks a number 1–20. You have limited guesses to find the secret *pumpkin seed*.

```python
secret = random.randint(1, 20)
attempts = 5
print("I'm thinking of a number between 1 and 20...")
won = False

while attempts > 0:
    guess = input(f"Take a guess ({attempts} left): ")
    if not guess.isdigit():
        print("Please enter a whole number.")
        continue
    guess = int(guess)
    if guess == secret:
        print("🎉 You found the magic pumpkin seed!")
        won = True
        break
    elif guess < secret:
        print("Too low... the crypt whispers 'higher'...")
    else:
        print("Too high... the ravens caw 'lower'...")
    attempts -= 1

if not won:
    print(f"💀 Alas! The number was {secret}. Better luck next haunt!")
```

## Game 2 — *Haunted Door* 🚪👻
Three doors. One hides a friendly ghost, the others hide jump‑scares. Pick wisely!

```python
doors = [1, 2, 3]
ghost_door = random.choice(doors)
choice = input("Choose a door (1, 2, or 3): ")
if choice.isdigit():
    choice = int(choice)
    if choice == ghost_door:
        print("👻 Boo! Friendly Ghost gives you candy! You win!")
    elif choice in doors:
        print("😱 Jumpscare! Maybe next time...")
    else:
        print("That door doesn't exist in this haunted house...")
else:
    print("You must choose a number.")
```

## Game 3 — *Tiny Text Adventure* 🗺️
You stand at a fork in the haunted woods. Choose a path and survive.

```python
print("You walk into the haunted woods and reach a fork: left or right?")
path = input("Type 'left' or 'right': ").strip().lower()

if path == "left":
    print("You find a cozy cabin with warm pumpkin pie. You rest safely. 🎃")
elif path == "right":
    print("A swarm of bats chases you to the graveyard...")
    action = input("Hide in a mausoleum or climb a tree? (mausoleum/tree): ").strip().lower()
    if action == "mausoleum":
        print("The door creaks shut behind you... but a secret tunnel leads home. You survive! 🏠")
    elif action == "tree":
        print("The branch snaps! You land in a pile of leaves. Startled, but safe. 🍂")
    else:
        print("Frozen by indecision, you become a statue in the graveyard. Spooky! 🗿")
else:
    print("You wander off the path and meet the Headless Programmer. He fixes your code. You live to debug another day!")
```

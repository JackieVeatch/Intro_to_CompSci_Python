## 🎯 Conditionals
Decide actions based on conditions.

```python
candies = 12
if candies > 20:
    print("Jackpot! Share with the coven.")
elif 10 <= candies <= 20:
    print("Nice haul. Trade for your favorites.")
else:
    print("Light the lanterns. We need more houses!")
```

**Comparisons & logic**

```python
age = 11
costume = "witch"
if (age >= 10 and age <= 13) and (costume == "witch" or costume == "vampire"):
    print("Eligible for the Junior Spook Contest!")
else:
    print("Maybe try the Pumpkin Parade.")
```

## 🔁 For Loops

```python
# Count ghosts
for g in range(1, 6):
    print("Ghost", g)
```

```python
houses = ["Elm St", "Maple Ave", "Pumpkin Ln"]
for street in houses:
    print("Knock knock at", street)
```

**Looping with indices**

```python
for i, street in enumerate(houses):
    print(f"{i}: {street}")
```

## 🔄 While Loops & Control

```python
candies = 0
while candies < 5:
    candies += 1
    if candies == 3:
        print("Found a full‑size bar! Skipping a happy dance...")
        continue
    if candies == 4:
        print("Uh oh, porch light off. Heading home early...")
        break
    print("Candy count:", candies)
```

## 🧪 Practice: Spooky FizzBuzz
Print numbers 1–30. For multiples of 3 print **Spook**, for multiples of 5 print **Boo**, for multiples of both print **SpookBoo**.

```python
for n in range(1, 31):
    if n % 3 == 0 and n % 5 == 0:
        print("SpookBoo")
    elif n % 3 == 0:
        print("Spook")
    elif n % 5 == 0:
        print("Boo")
    else:
        print(n)
```

In a previous class, we learned that creating functions are a convenient way to organize our code, so that we do not have to write out the same piece of code multiple times, but rather call upon a function. <br>

```python
# --- Convert Celsius to Fahrenheit (no functions yet) ---

c1 = 0
c2 = 10
c3 = 25
c4 = 37
c5 = 100

f1 = (c1 * 9/5) + 32
f2 = (c2 * 9/5) + 32
f3 = (c3 * 9/5) + 32
f4 = (c4 * 9/5) + 32
f5 = (c5 * 9/5) + 32

print(f"{c1}°C = {f1}°F")
print(f"{c2}°C = {f2}°F")
print(f"{c3}°C = {f3}°F")
print(f"{c4}°C = {f4}°F")
print(f"{c5}°C = {f5}°F")
```

The above code works! But there are a few ways we can optimize it. The code repeats a similar calculation several times. What if we made this repeated calculation into a function? <br>

Functions begin with `def` followed by the name of the function and the function inputs: `name(inputs)`. The code that is inside the function is indented to the right. Where the function ends is where the code is no longer indented.

```python
def c_to_f(celsius):
    """Convert Celsius to Fahrenheit."""
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

f0 = c_to_f(0)
print(f0)

f1 = c_to_f(10)
print(f1)

f2 = c_to_f(25)
print(f2)

f3 = c_to_f(37)
print(f3)

f4 = c_to_f(100)
print(f4)
```

Now we're using a function! But we still have a lot of repeating code. Finish the loop below to loop through each value of Celcius and convert them to Fahrenheit.

```python
# Now you can reuse the function for any number of temperatures:
temps_c = [0, 10, 25, 37, 100]

for c in temps_c:
    f = c_to_f(c)
    print(f"{c}°C = {f:.1f}°F")
```

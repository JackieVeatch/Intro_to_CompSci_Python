## 2. `if / elif / else`: Multiple Branches

Sometimes we want more than one possibility. That's where `if / elif / else` comes in.

Example: grading based on a score.

```python
score = int(input("Enter your test score (0-100): "))

if score >= 90:
    print("You got an A!")
elif score >= 80:
    print("You got a B.")
elif score >= 70:
    print("You got a C.")
else:
    print("Keep trying, you can improve next time!")
```

## 3. Embedded (Nested) Conditionals

Now for the new idea: **embedded** or **nested** conditionals.

This is when you put an `if` statement **inside** another `if` (or `elif` or `else`).

Think of it like a **decision tree**:

- First question: Are you a student?
  - If **yes** → Next question: Are you in high school or college?
  - If **no** → Different message.

In code:

```python
is_student = input("Are you a student? (yes/no) ").lower()

if is_student == "yes":
    level = input("Are you in high school or college? ").lower()

    if level == "high school":
        print("High school student discount applied!")
    elif level == "college":
        print("College student discount applied!")
    else:
        print("Student level not recognized, but hi anyway!")
else:
    print("No student discount, but welcome!")
```

### Visualizing the Branching Logic Tree

For the code above, the logic tree looks like this:

- Are you a student?
  - **yes** → Ask: "Are you in high school or college?"
    - **high school** → `High school student discount applied!`
    - **college** → `College student discount applied!`
    - **anything else** → `Student level not recognized, but hi anyway!`
  - **no** → `No student discount, but welcome!`

Each branch represents a different path through your code.

## 1. What is a class?

A **class** is a blueprint for creating objects.

A class describes:
- what data an object should store (an object is a container for data (attributes) that is created from a class)
- what actions an object should be able to perform

For example, imagine a `Dog` class.

A dog object might store:
- a name
- an age
- a breed

And it might be able to do things like:
- bark
- play fetch

The class is the template. Each actual dog you create from that template is an object.


## 2. Why use classes?

Classes are useful when you want to group related data and behavior together.

Without classes, you might store information in separate variables:

```python
dog1_name = "Milo"
dog1_age = 3
dog1_breed = "Beagle"
```

That works for a tiny program, but it becomes messy if you have lots of dogs, students, video game characters, bank accounts, or sensors.

With a class, you can keep everything organized in one object.

### Main reasons to use classes

1. **Organization** – related data and actions stay together.
2. **Reusability** – you can make many objects from one class.
3. **Readability** – your code becomes easier to understand.
4. **Modeling real things** – classes are great for representing people, pets, cars, books, game characters, and more.


## 3. Your first class

Here is a very simple class:


````python
class Dog:
    def bark(self):
        print("Woof!")


my_dog = Dog()
my_dog.bark()
````

### What happened here?

- `class Dog:` creates a class called `Dog`
- `def bark(self):` defines a method
- `my_dog = Dog()` creates an object from the class
- `my_dog.bark()` tells that object to run its `bark` method

A method is just a function that belongs to a class.


## 4. What is `self`?

In a class, `self` refers to the specific object using the method.

You can think of `self` as meaning:

> "this particular object"

When you write:

```python
my_dog.bark()
```

Python automatically passes `my_dog` into the method as `self`.

That lets each object keep track of its own data.


## 5. Adding attributes with `__init__`

Most classes use a special method called `__init__()`.

This method runs when a new object is created. It is used to set up the object's starting data. I'm pretty sure "init" is short for initiate.


````python
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        print(f"{self.name} says woof!")

    def birthday(self):
        self.age += 1
        print(f"Happy birthday, {self.name}! You are now {self.age}.")


dog1 = Dog("Milo", 3, "Beagle")
dog2 = Dog("Luna", 5, "Border Collie")

print(dog1.name)
print(dog2.breed)
dog1.bark()
dog2.birthday()
````

### Vocabulary check

- **Class**: the blueprint
- **Object**: a specific instance of a class
- **Attribute**: data stored in an object, like `name` or `age`
- **Method**: an action the object can perform, like `bark()`
- **Instance**: another word for an object made from a class


## 6. Example: a Video Game Character class

Classes are especially useful when you want to model something with both data and actions.


````python
class GameCharacter:
    def __init__(self, name, health, power):
        self.name = name
        self.health = health
        self.power = power

    def attack(self):
        print(f"{self.name} attacks with power {self.power}!")

    def take_damage(self, amount):
        self.health -= amount
        print(f"{self.name} takes {amount} damage and now has {self.health} health.")

    def heal(self, amount):
        self.health += amount
        print(f"{self.name} heals {amount} points and now has {self.health} health.")


hero = GameCharacter("Aria", 100, 18)
villain = GameCharacter("Blaze", 120, 15)

hero.attack()
villain.take_damage(20)
hero.heal(10)
````

## 7. Why not just use functions?

You absolutely *can* solve many problems using just functions and variables.

But classes become helpful when:
- you have many similar things to keep track of
- each thing has its own data
- each thing can perform similar actions

For example:
- a classroom full of students
- a game with many characters
- a library with many books
- a weather app tracking many cities

Classes help you avoid repeated code.

## 12. Slightly more advanced idea: changing object state

One powerful part of classes is that objects can change over time.

For example, a `Timer` object might count up, or a `BankAccount` object might gain or lose money.

That means an object is not just storing information — it can also update itself.


````python
class Timer:
    def __init__(self):
        self.seconds = 0

    def tick(self):
        self.seconds += 1

    def show_time(self):
        print(f"Timer: {self.seconds} seconds")


t = Timer()
t.show_time()
t.tick()
t.tick()
t.show_time()
````

## 13. Common mistakes to watch for

### Forgetting `self`
Inside a class method, you usually need `self` as the first parameter.

### Forgetting `self.` before attributes
Inside methods, `name` is different from `self.name`.

### Indentation problems
Everything inside the class and methods must be indented correctly.

### Calling a method without parentheses
`dog.bark` is not the same as `dog.bark()`.

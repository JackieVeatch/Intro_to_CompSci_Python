## 1) Vectors

A **vector** is an ordered list of numbers.

Examples:
- 2D vector: $\begin{bmatrix} x \\ y \end{bmatrix}$
- 3D vector: $\begin{bmatrix} x \\ y \\ z \end{bmatrix}$

You can think of a vector as:
- an **arrow** in space
- a **direction + magnitude**
- a bundle of measurements/features

### Creating vectors in NumPy
We’ll use 1D arrays for vectors.

```python
v = np.array([3, 4])
w = np.array([-1, 2])

print("v =", v)
print("w =", w)
```

### Vector addition and scaling
- Add vectors component-by-component
- Multiply a vector by a number (a scalar) to stretch or shrink it

```python
print("v + w =", v + w)
print("2*v  =", 2*v)
print("0.5*w=", 0.5*w)
```

### Vector magnitude (length)
The length of a vector $v$ is:
$$\|v\| = \sqrt{x^2 + y^2}$$

NumPy has `np.linalg.norm` for this.

```python
print("||v|| =", np.linalg.norm(v))
print("||w|| =", np.linalg.norm(w))
```

### Dot product (how aligned are two vectors?)
The dot product is:
$$v \cdot w = x_1 x_2 + y_1 y_2$$

- If $v\cdot w$ is **positive**, they point mostly the same way
- If it’s **0**, they’re perpendicular
- If it’s **negative**, they point mostly opposite directions

```python
dot = np.dot(v, w)
print("v · w =", dot)
```

### Quick visualization of vectors in 2D
Let’s draw arrows from the origin.

```python
def plot_vectors(vectors, labels=None, title="Vectors"):
    plt.figure(figsize=(6,6))
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    ax.axhline(0)
    ax.axvline(0)

    max_len = 1
    for i, vec in enumerate(vectors):
        x, y = vec
        max_len = max(max_len, abs(x), abs(y))
        lbl = labels[i] if labels else None
        ax.arrow(0, 0, x, y, head_width=0.2, length_includes_head=True)
        if lbl:
            ax.text(x, y, f" {lbl}")

    lim = max_len + 1
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_title(title)
    plt.show()

plot_vectors([v, w], labels=["v", "w"], title="v and w")
```

## 2) Matrices

A **matrix** is a rectangular grid of numbers.

A 2×2 matrix looks like:
$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$

In this notebook, we’ll mostly focus on **2×2 matrices** because we can visualize them easily.

### Creating matrices in NumPy

```python
A = np.array([[2, 1],
              [1, 2]])

B = np.array([[0, -1],
              [1,  0]])

print("A=\n", A)
print("\nB=\n", B)
```

### Matrix addition and scalar multiplication
Works like you’d expect: add entries in matching positions.

```python
print("A + A =\n", A + A)
print("3*A =\n", 3*A)
```

### Matrix multiplication (important!)
Matrix multiplication is **not** done entry-by-entry.

In NumPy:
- Use `A @ v` for matrix-vector multiplication
- Use `A @ B` for matrix-matrix multiplication

⚠️ Avoid `*` for matrix multiplication. `*` does element-by-element multiplication.

```python
v = np.array([3, 1])

print("A @ v =", A @ v)
print("B @ v =", B @ v)
print("\nA @ B =\n", A @ B)
print("\nB @ A =\n", B @ A)
```

Notice that **A @ B** is generally different from **B @ A**.

That means matrix multiplication is usually **not commutative**.

## 3) Matrices as transformations

A 2×2 matrix can act on a 2D vector and produce a new vector.

This is like a machine:
$$\text{input vector} \rightarrow \text{matrix} \rightarrow \text{output vector}$$

Examples:
- Rotations
- Stretches/squeezes
- Reflections
- Shears

### Helper: plot a transformation of many points
We’ll transform a grid of points and see what happens.

```python
def plot_transformation(M, title="Transformation", grid_lim=2, step=0.5):
    # Create grid points
    xs = np.arange(-grid_lim, grid_lim + step, step)
    ys = np.arange(-grid_lim, grid_lim + step, step)
    pts = np.array([(x, y) for x in xs for y in ys])

    transformed = (M @ pts.T).T

    plt.figure(figsize=(7,7))
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    ax.axhline(0)
    ax.axvline(0)

    ax.scatter(pts[:,0], pts[:,1], label='original')
    ax.scatter(transformed[:,0], transformed[:,1], label='transformed')

    ax.set_title(title)
    ax.legend()
    plt.show()
```

### Example: rotation by 90°
The matrix
$$R = \begin{bmatrix}0 & -1 \\ 1 & 0\end{bmatrix}$$
rotates vectors 90° counterclockwise.

```python
R90 = np.array([[0, -1],
                [1,  0]])

plot_transformation(R90, title="Rotation by 90°")
```

### Example: stretch in x direction
$$S = \begin{bmatrix}2 & 0 \\ 0 & 1\end{bmatrix}$$
doubles x values but leaves y alone.

```python
S = np.array([[2, 0],
              [0, 1]])

plot_transformation(S, title="Stretch x by 2")
```

## 4) Eigenvectors and eigenvalues

When a matrix transforms vectors, **most vectors change direction**.

But some special vectors keep their direction (they may stretch or shrink).

A vector **v** is an **eigenvector** of matrix **A** if:
$$A\,v = \lambda v$$

- **v** is an eigenvector
- **λ (lambda)** is an eigenvalue

Meaning:
- The matrix transforms **v** into the same direction
- It scales it by **λ**

### Why do we care?
Eigenvectors show the matrix’s **preferred directions**.
They show up everywhere: physics, computer graphics, Google’s PageRank, data science (PCA), etc.

### Compute eigenvalues/eigenvectors in NumPy
Use `np.linalg.eig(A)`.

```python
A = np.array([[2, 1],
              [1, 2]])

eigvals, eigvecs = np.linalg.eig(A)

print("A =\n", A)
print("\nEigenvalues:", eigvals)
print("\nEigenvectors (columns):\n", eigvecs)
```

### Check the eigenvector equation
If `v` is an eigenvector and `lam` is its eigenvalue, then `A @ v` should equal `lam * v`.

Let’s test each eigenvector.

```python
for i in range(len(eigvals)):
    lam = eigvals[i]
    v = eigvecs[:, i]
    left = A @ v
    right = lam * v
    print(f"\nEigenpair {i}:")
    print("lambda =", lam)
    print("v      =", v)
    print("A @ v  =", left)
    print("lam*v  =", right)
    print("difference =", left - right)
```

Small differences are just floating point rounding.

### Visualizing eigenvectors
We’ll plot the eigenvectors as arrows.

```python
vec1 = eigvecs[:, 0]
vec2 = eigvecs[:, 1]

plot_vectors([vec1, vec2], labels=["e1", "e2"], title="Eigenvectors of A")
```

### What does A do to its eigenvectors?
Let’s compare each eigenvector with its transformed version `A@v`.

If it’s truly an eigenvector, the transformed arrow should lie exactly on the same line.

```python
def compare_vector_and_transform(M, v, title="Compare v and M@v"):
    mv = M @ v
    plot_vectors([v, mv], labels=["v", "M@v"], title=title)

for i in range(2):
    lam = eigvals[i]
    v = eigvecs[:, i]
    compare_vector_and_transform(A, v, title=f"Eigenvector {i}: lambda={lam:.3f}")
```

## 5) A geometric way to understand eigenvectors

Try this idea:
1. Draw a bunch of vectors in different directions.
2. Apply the matrix to all of them.
3. Look for directions that don’t rotate (they only stretch).

We’ll do a quick experiment: create many unit vectors around the circle, transform them, and compare angles.

```python
def unit_vectors_on_circle(n=36):
    angles = np.linspace(0, 2*np.pi, n, endpoint=False)
    return np.column_stack([np.cos(angles), np.sin(angles)]), angles

def angle_of(v):
    return np.arctan2(v[1], v[0])

pts, angles = unit_vectors_on_circle(72)

rot_changes = []
for u in pts:
    Mu = A @ u
    rot_changes.append(angle_of(Mu) - angle_of(u))

rot_changes = np.unwrap(np.array(rot_changes))

plt.figure(figsize=(7,4))
plt.plot(angles, rot_changes)
plt.axhline(0)
plt.title("Change in angle after applying A (where change≈0 suggests eigenvector directions)")
plt.xlabel("original angle (radians)")
plt.ylabel("angle change (radians)")
plt.show()
```

Places where the curve crosses 0 correspond to directions that don’t rotate.
Those directions are eigenvector directions.
